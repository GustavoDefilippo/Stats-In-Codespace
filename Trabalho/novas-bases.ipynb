# Metadados - Rede de Rodovias Federais Brasileiras (SNV)

## Dicionário de Dados - Sistema Nacional de Viação

| Campo | Tipo | Formato | Descrição |
|---|---|---|---|
| BR | Numérico | NN | Número da rodovia federal (ex: BR-010, BR-116). Identifica a rodovia de forma única no sistema federal. |
| UF | Texto | XX | Sigla da Unidade Federativa onde o trecho está localizado (ex: DF, GO, MA, PA). Formato: 2 caracteres. |
| Tipo de Trecho | Texto | Livre | Classificação do tipo de trecho da rodovia (ex: Eixo Principal, Eixo Secundário, Ligação). Define a importância da rodovia na rede. |
| Desc Coinc | Texto | Livre | Descrição de coincidência/sobreposição de rodovias. Campo geralmente vazio (-) quando não há coincidência com outras vias. |
| Código | Alfanumérico | NNNAANNNNNN | Código único do trecho (ex: 010BDF0010). Formato: número BR + letra + UF + sequência numérica. |
| Local de Início | Texto | Livre | Descrição textual do local de origem do trecho (ex: "ENTR BR-020(A)/030(A)/450/DF-001 (BRASÍLIA)"). |
| Local de Fim | Texto | Livre | Descrição textual do local de destino do trecho (ex: "ENTR DF-440"). |
| km inicial | Numérico | NNN,N | Quilômetro inicial do trecho na rodovia (ex: 0, 2.4, 6). Indicador de posição inicial do segmento. |
| km final | Numérico | NNN,N | Quilômetro final do trecho na rodovia (ex: 2.4, 6, 8.3). Indicador de posição final do segmento. |
| Extensão | Numérico | NNN,N | Extensão total do trecho em quilômetros (km final - km inicial). Calculada automaticamente. |
| Superfície Federal | Texto | Código | Tipo de superfície de pavimentação da rodovia federal (ex: DUP = Duplicada, PAV = Pavimentada, PLA = Planejada). |
| OBRAS | Texto | Código | Indicação de obras em execução ou planejadas no trecho (ex: EOD = Em Ordem de Execução; geralmente vazio). |
| Federal Coincidente | Texto | Livre | Código(s) de outras rodovias federais que coincidem com este trecho (ex: "030BDF0010;020BDF0010"). Separadas por ponto-e-vírgula. |
| Administração | Texto | Livre | Entidade administrativa responsável pela manutenção (ex: "Convênio de Administração", "Distrital", "Estadual", "Federal"). |
| Ato Legal | Texto | Código | Referência legal ou instrumento normativo que regulamenta o trecho (geralmente vazio neste dataset). |
| Estadual Coincidente | Texto | Livre | Código(s) de rodovias estaduais que coincidem com este trecho (ex: "DF-345", "GO-118"). |
| Superfície Est. Coincidente | Texto | Código | Tipo de superfície de pavimentação das rodovias estaduais coincidentes (ex: PAV = Pavimentada). |
| Jurisdição | Texto | Livre | Entidade responsável pela jurisdição do trecho (ex: "Federal", "Distrital", "Estadual"). Pode difereir da Administração. |
| Superfície | Texto | Código | Tipo de superfície do trecho (ex: PAV = Pavimentada, PLA = Planejada, DUP = Duplicada). Resumido/principal. |
| Unidade Local | Texto | Livre | Divisão administrativa/regional responsável pelo trecho (ex: "Brasília", "Imperatriz", "São Miguel do Guamá"). Unidade operacional do DNIT. |

## Notas sobre a Estrutura

- **Chave Primária**: `Código` (identifica unicamente cada trecho)
- **Chave Composta Alternativa**: `BR + UF + km_inicial` (para rastreabilidade geográfica)
- **Campos de Relacionamento**: `Federal Coincidente`, `Estadual Coincidente` (múltiplos valores separados por `;`)
- **Valores Vazios**: Representados por `-` ou células vazias; devem ser padronizados como `NA` no carregamento
- **Formato de Números**: Vírgula (`,`) como separador decimal (padrão brasileiro)
- **Granularidade**: Dados em nível de **trecho de rodovia** (segmento entre dois pontos de referência)

```r
################################################################################
#       CARGA E LIMPEZA DO DATASET SNV - SISTEMA NACIONAL DE VIAÇÃO            #
################################################################################

# Carregar bibliotecas necessárias
library(tidyverse)
library(readr)

################################################################################
#                         CARGA DO ARQUIVO CSV                                #
################################################################################

# Caminho do arquivo
caminho_arquivo <- "Trabalho/bases-de-dados/definicao-de-pavimento/SNV_202101Aa.csv"

# Carga com especificação de tipos e tratamento de valores ausentes
df_snv <- read_delim(
  caminho_arquivo,
  delim = ";",
  locale = locale(
    encoding = "UTF-8",
    decimal_mark = ","  # Padrão brasileiro: vírgula como separador decimal
  ),
  na = c("", "-", "NA"),  # Valores a serem convertidos em NA
  col_types = cols(
    BR = col_integer(),                    # Número da rodovia (ex: 10)
    UF = col_character(),                  # Sigla UF (ex: DF, GO, MA, PA)
    `Tipo de trecho` = col_character(),    # Classificação do trecho
    `Desc Coinc` = col_character(),        # Descrição de coincidência
    Código = col_character(),              # Código único do trecho (chave primária)
    `Local de Início` = col_character(),   # Descrição local início
    `Local de Fim` = col_character(),      # Descrição local fim
    `km inicial` = col_double(),           # Km inicial (numérico com decimais)
    `km final` = col_double(),             # Km final (numérico com decimais)
    Extensão = col_double(),               # Extensão em km (numérico)
    `Superfície Federal` = col_character(),# Tipo superfície: DUP, PAV, PLA
    OBRAS = col_character(),               # Indicação obras (EOD, etc)
    `Federal Coincidente` = col_character(),  # Rodovias federais coincidentes
    Administração = col_character(),       # Entidade administradora
    `Ato legal` = col_character(),         # Referência legal
    `Estadual Coincidente` = col_character(), # Rodovias estaduais coincidentes
    `Superfície Est. Coincidente` = col_character(), # Superfície estadual
    Jurisdição = col_character(),          # Entidade com jurisdição
    Superfície = col_character(),          # Superfície principal
    `Unidade Local` = col_character()      # Divisão administrativa/regional
  ),
  trim_ws = TRUE  # Remove espaços em branco nas extremidades
)

################################################################################
#                    VISUALIZAÇÃO E VALIDAÇÃO INICIAL                         #
################################################################################

cat("\n=== DIMENSÕES DO DATASET ===\n")
cat("Linhas:", nrow(df_snv), "\n")
cat("Colunas:", ncol(df_snv), "\n\n")

cat("=== ESTRUTURA DOS DADOS ===\n")
str(df_snv)

cat("\n=== PRIMEIRAS 10 LINHAS ===\n")
print(head(df_snv, 10))

################################################################################
#                    ANÁLISE DE VALORES AUSENTES (NA)                          #
################################################################################

cat("\n=== CONTAGEM DE VALORES AUSENTES POR COLUNA ===\n")
valores_ausentes <- colSums(is.na(df_snv))
print(valores_ausentes[valores_ausentes > 0])

cat("\n=== PERCENTUAL DE VALORES AUSENTES POR COLUNA ===\n")
percentual_na <- round(100 * colMeans(is.na(df_snv)), 2)
print(percentual_na[percentual_na > 0])

################################################################################
#                    ANÁLISE EXPLORATÓRIA DOS DADOS                            #
################################################################################

cat("\n=== TIPOS DE DADOS CARREGADOS ===\n")
print(sapply(df_snv, class))

cat("\n=== VALORES ÚNICOS - VARIÁVEIS CATEGÓRICAS PRINCIPAIS ===\n")
cat("\nBR (Rodovias Federais):\n")
print(sort(unique(df_snv$BR)))

cat("\nUF (Unidades Federativas):\n")
print(sort(unique(df_snv$UF)))

cat("\nTipo de Trecho:\n")
print(unique(df_snv$`Tipo de trecho`))

cat("\nSuperfície Federal (categorias):\n")
print(unique(df_snv$`Superfície Federal`))

cat("\nAdministração (responsáveis):\n")
print(unique(df_snv$Administração))

cat("\nJurisdição:\n")
print(unique(df_snv$Jurisdição))

################################################################################
#                    ESTATÍSTICAS DAS VARIÁVEIS NUMÉRICAS                      #
################################################################################

cat("\n=== ESTATÍSTICAS DESCRITIVAS - QUILOMETRAGEM ===\n")
df_snv %>%
  summarise(
    `km_inicial (mín)` = min(`km inicial`, na.rm = TRUE),
    `km_inicial (máx)` = max(`km inicial`, na.rm = TRUE),
    `km_inicial (média)` = round(mean(`km inicial`, na.rm = TRUE), 2),
    `km_final (mín)` = min(`km final`, na.rm = TRUE),
    `km_final (máx)` = max(`km final`, na.rm = TRUE),
    `km_final (média)` = round(mean(`km final`, na.rm = TRUE), 2),
    `Extensão (mín)` = min(Extensão, na.rm = TRUE),
    `Extensão (máx)` = max(Extensão, na.rm = TRUE),
    `Extensão (média)` = round(mean(Extensão, na.rm = TRUE), 2),
    `Extensão (total)` = round(sum(Extensão, na.rm = TRUE), 2)
  ) %>%
  print()

################################################################################
#                    RESUMO POR UNIDADE FEDERATIVA                             #
################################################################################

cat("\n=== RESUMO EXECUTIVO POR ESTADO ===\n")
resumo_uf <- df_snv %>%
  group_by(UF) %>%
  summarise(
    Total_Trechos = n(),
    Extensao_Total_km = round(sum(Extensão, na.rm = TRUE), 2),
    Extensao_Media_km = round(mean(Extensão, na.rm = TRUE), 2),
    Trechos_NA = sum(is.na(Extensão)),
    .groups = 'drop'
  ) %>%
  arrange(desc(Extensao_Total_km))

print(resumo_uf)

################################################################################
#                    RESUMO POR TIPO DE SUPERFÍCIE                             #
################################################################################

cat("\n=== RESUMO POR SUPERFÍCIE FEDERAL ===\n")
resumo_superficie <- df_snv %>%
  group_by(`Superfície Federal`) %>%
  summarise(
    Total_Trechos = n(),
    Extensao_Total_km = round(sum(Extensão, na.rm = TRUE), 2),
    Extensao_Media_km = round(mean(Extensão, na.rm = TRUE), 2),
    .groups = 'drop'
  ) %>%
  arrange(desc(Extensao_Total_km))

print(resumo_superficie)

################################################################################
#                    VALIDAÇÃO DE INTEGRIDADE DOS DADOS                        #
################################################################################

cat("\n=== VERIFICAÇÃO DE INTEGRIDADE ===\n")

# Verificar se há códigos duplicados (devem ser únicos)
codigos_duplicados <- df_snv %>%
  group_by(Código) %>%
  filter(n() > 1) %>%
  nrow()
cat("Códigos duplicados:", codigos_duplicados, "\n")

# Verificar consistência: km_final deve ser >= km_inicial
inconsistencias <- df_snv %>%
  filter(`km final` < `km inicial`) %>%
  nrow()
cat("Inconsistências (km_final < km_inicial):", inconsistencias, "\n")

# Verificar consistência: Extensão deve igualar (km_final - km_inicial)
tolerancia <- 0.01  # Tolerância para diferenças de arredondamento
inconsistencias_extensao <- df_snv %>%
  mutate(extensao_calculada = `km final` - `km inicial`) %>%
  filter(abs(Extensão - extensao_calculada) > tolerancia) %>%
  nrow()
cat("Inconsistências de extensão:", inconsistencias_extensao, "\n")

################################################################################
#                    CONFIRMAÇÃO DE CARGA FINALIZADA                           #
################################################################################

cat("\n✓ Dataset SNV carregado com sucesso!\n")
cat("   Total de trechos:", nrow(df_snv), "\n")
cat("   Extensão total de rodovias federais:", 
    round(sum(df_snv$Extensão, na.rm = TRUE), 2), "km\n")
cat("   Estados cobertos:", n_distinct(df_snv$UF), "\n")
cat("   Rodovias federais:", n_distinct(df_snv$BR), "\n\n")

# Salvar dados processados em RData para uso posterior
save(df_snv, file = "Trabalho/bases-de-dados/definicao-de-pavimento/df_snv.RData")
cat("✓ Dataset salvo em RData para carregamento futuro\n")
```

---

## Notas sobre a Estrutura

- **Chave Primária**: `id` (identifica unicamente cada acidente registrado)
- **Chave Geográfico-Temporal Composta**: `uf + br + data_inversa + horario` (rastreabilidade de localização e tempo)
- **Campos Multivalorados**: `tipo_acidente`, `tracado_via` (separados por `;`)
- **Valores Ausentes**: Representados como `NA` ou células vazias; devem ser padronizados no carregamento
- **Formato de Números**: Vírgula (`,`) como separador decimal (padrão brasileiro)
- **Codificação Original**: ISO-8859-1 (deve ser convertida para UTF-8 ao carregar)
- **Granularidade**: Dados em nível de **registro individual de acidente** (não agregados)
- **Campos Derivados**: `feridos` = `feridos_leves + feridos_graves`; `pessoas` = `mortos + feridos + ilesos + ignorados`

---

## Validações Esperadas

| Validação | Descrição |
|---|---|
| IDs únicos | Não deve haver `id` duplicado |
| Coerência de vítimas | `feridos` deve igualar `feridos_leves + feridos_graves` |
| Coerência de pessoas | `pessoas` deve ser >= soma de componentes (mortos, feridos, ilesos) |
| Coordenadas geográficas | Latitude entre -33° e +5°; Longitude entre -74° e -34° (Brasil) |
| Dados temporais | `data_inversa` em formato válido; `horario` entre 00:00:00 e 23:59:59 |

## Contexto de Uso

**Fonte oficial**: [Portal de Dados Abertos da PRF](https://www.gov.br/prf/pt-br/acesso-a-informacao/dados-abertos/dados-abertos-da-prf)

**Período**: Dados coletados a partir de 2024-01-01

**Unidade de análise**: Registro individual de acidente rodoviário em rodovia federal brasileira

**Aplicações**: Análise de padrões de segurança viária, identificação de pontos críticos, estudos de causalidade de acidentes, planejamento de intervenções de redução de sinistralidade

```r
################################################################################
#    CARGA E LIMPEZA DO DATASET PRF - ACIDENTES EM RODOVIAS FEDERAIS           #
################################################################################

# Carregar bibliotecas necessárias
library(tidyverse)
library(readr)

################################################################################
#                    CARGA DO ARQUIVO CSV COM CONVERSÃO UTF-8                  #
################################################################################

# Caminho do arquivo
caminho_arquivo <- "Trabalho/bases-de-dados/acidentes/datatran2024.csv"

# Carga com especificação de tipos e tratamento de valores ausentes
# Convertendo de ISO-8859-1 para UTF-8 durante a leitura
df_acidentes_prf <- read_delim(
  caminho_arquivo,
  delim = ";",
  locale = locale(
    encoding = "ISO-8859-1",  # Encoding original do arquivo PRF
    decimal_mark = ","         # Padrão brasileiro: vírgula como separador decimal
  ),
  na = c("", "NA", "NA "),     # Valores a serem convertidos em NA
  col_types = cols(
    id = col_integer(),                      # ID único do acidente
    data_inversa = col_date(format = "%Y-%m-%d"),  # Data no formato inverso (AAAA-MM-DD)
    dia_semana = col_character(),            # Dia da semana (segunda-feira, terça-feira, etc)
    horario = col_time(format = "%H:%M:%S"), # Hora do acidente (HH:MM:SS)
    uf = col_character(),                    # Sigla da UF (ES, PI, BA, etc)
    br = col_integer(),                      # Número da rodovia federal (101, 343, 116, etc)
    km = col_double(),                       # Quilômetro onde ocorreu o acidente
    municipio = col_character(),             # Município do acidente
    causa_acidente = col_character(),        # Causa principal do acidente
    tipo_acidente = col_character(),         # Tipo (colisão, tombamento, saída de pista, etc)
    classificacao_acidente = col_character(),# Classificação (Sem Vítimas, Com Vítimas Feridas, etc)
    fase_dia = col_character(),              # Fase do dia (Plena Noite, Amanhecer, Pleno dia, etc)
    sentido_via = col_character(),           # Sentido (Crescente, Decrescente)
    condicao_metereologica = col_character(),# Condição meteorológica (Céu Claro, Chuva, Nublado, etc)
    tipo_pista = col_character(),            # Tipo de pista (Simples, Dupla)
    tracado_via = col_character(),           # Traçado (Reta, Curva, Aclive, Declive, etc)
    uso_solo = col_character(),              # Uso do solo (Sim, Não)
    pessoas = col_integer(),                 # Total de pessoas envolvidas
    mortos = col_integer(),                  # Número de óbitos
    feridos_leves = col_integer(),           # Número de feridos leves
    feridos_graves = col_integer(),          # Número de feridos graves
    ilesos = col_integer(),                  # Número de ilesos
    ignorados = col_integer(),               # Número de ignorados
    feridos = col_integer(),                 # Total de feridos (leves + graves)
    veiculos = col_integer(),                # Número de veículos envolvidos
    latitude = col_double(),                 # Coordenada latitude (formato decimal)
    longitude = col_double(),                # Coordenada longitude (formato decimal)
    regional = col_character(),              # Superintendência Regional da PRF
    delegacia = col_character(),             # Código da delegacia
    uop = col_character()                    # Unidade Operacional da PRF
  ),
  trim_ws = TRUE  # Remove espaços em branco nas extremidades
)

# Converter explicitamente para UTF-8 e garantir codificação correta
Encoding(df_acidentes_prf$causa_acidente) <- "UTF-8"
Encoding(df_acidentes_prf$tipo_acidente) <- "UTF-8"
Encoding(df_acidentes_prf$classificacao_acidente) <- "UTF-8"
Encoding(df_acidentes_prf$condicao_metereologica) <- "UTF-8"
Encoding(df_acidentes_prf$municipio) <- "UTF-8"
Encoding(df_acidentes_prf$tracado_via) <- "UTF-8"

################################################################################
#                    VISUALIZAÇÃO E VALIDAÇÃO INICIAL                         #
################################################################################

cat("\n=== DIMENSÕES DO DATASET ===\n")
cat("Linhas:", nrow(df_acidentes_prf), "\n")
cat("Colunas:", ncol(df_acidentes_prf), "\n\n")

cat("=== ESTRUTURA DOS DADOS ===\n")
str(df_acidentes_prf)

cat("\n=== PRIMEIRAS 10 LINHAS ===\n")
print(head(df_acidentes_prf, 10))

################################################################################
#                    ANÁLISE DE VALORES AUSENTES (NA)                          #
################################################################################

cat("\n=== CONTAGEM DE VALORES AUSENTES POR COLUNA ===\n")
valores_ausentes <- colSums(is.na(df_acidentes_prf))
print(valores_ausentes[valores_ausentes > 0])

cat("\n=== PERCENTUAL DE VALORES AUSENTES POR COLUNA ===\n")
percentual_na <- round(100 * colMeans(is.na(df_acidentes_prf)), 2)
print(percentual_na[percentual_na > 0])

################################################################################
#                    ANÁLISE EXPLORATÓRIA DOS DADOS                            #
################################################################################

cat("\n=== TIPOS DE DADOS CARREGADOS ===\n")
print(sapply(df_acidentes_prf, class))

cat("\n=== DISTRIBUIÇÃO POR UNIDADE FEDERATIVA ===\n")
print(sort(unique(df_acidentes_prf$uf)))

cat("\n=== DISTRIBUIÇÃO POR RODOVIA FEDERAL ===\n")
print(sort(unique(df_acidentes_prf$br)))

cat("\n=== VALORES ÚNICOS - VARIÁVEIS CATEGÓRICAS PRINCIPAIS ===\n")

cat("\nDia da Semana:\n")
print(unique(df_acidentes_prf$dia_semana))

cat("\nTipo de Acidente:\n")
print(unique(df_acidentes_prf$tipo_acidente))

cat("\nClassificação do Acidente:\n")
print(unique(df_acidentes_prf$classificacao_acidente))

cat("\nFase do Dia:\n")
print(unique(df_acidentes_prf$fase_dia))

cat("\nCondição Meteorológica:\n")
print(unique(df_acidentes_prf$condicao_metereologica))

cat("\nTipo de Pista:\n")
print(unique(df_acidentes_prf$tipo_pista))

################################################################################
#                    ESTATÍSTICAS DAS VARIÁVEIS NUMÉRICAS                      #
################################################################################

cat("\n=== ESTATÍSTICAS DESCRITIVAS - VÍTIMAS E VEÍCULOS ===\n")
df_acidentes_prf %>%
  summarise(
    `Total de Acidentes` = n(),
    `Pessoas Envolvidas (total)` = sum(pessoas, na.rm = TRUE),
    `Pessoas Envolvidas (média)` = round(mean(pessoas, na.rm = TRUE), 2),
    `Total de Óbitos` = sum(mortos, na.rm = TRUE),
    `Total de Feridos` = sum(feridos, na.rm = TRUE),
    `Feridos Leves` = sum(feridos_leves, na.rm = TRUE),
    `Feridos Graves` = sum(feridos_graves, na.rm = TRUE),
    `Total de Veículos` = sum(veiculos, na.rm = TRUE),
    `Veículos por Acidente (média)` = round(mean(veiculos, na.rm = TRUE), 2)
  ) %>%
  print()

################################################################################
#                    ESTATÍSTICAS - QUILOMETRAGEM                              #
################################################################################

cat("\n=== ESTATÍSTICAS DESCRITIVAS - QUILOMETRAGEM ===\n")
df_acidentes_prf %>%
  summarise(
    `km (mín)` = min(km, na.rm = TRUE),
    `km (máx)` = max(km, na.rm = TRUE),
    `km (média)` = round(mean(km, na.rm = TRUE), 2),
    `km (mediana)` = round(median(km, na.rm = TRUE), 2),
    `km (desvio padrão)` = round(sd(km, na.rm = TRUE), 2)
  ) %>%
  print()

################################################################################
#                    RESUMO POR UNIDADE FEDERATIVA                             #
################################################################################

cat("\n=== RESUMO EXECUTIVO POR ESTADO ===\n")
resumo_uf <- df_acidentes_prf %>%
  group_by(uf) %>%
  summarise(
    Total_Acidentes = n(),
    Total_Obitos = sum(mortos, na.rm = TRUE),
    Total_Feridos = sum(feridos, na.rm = TRUE),
    Total_Pessoas = sum(pessoas, na.rm = TRUE),
    Media_Pessoas_Acidente = round(mean(pessoas, na.rm = TRUE), 2),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total_Acidentes))

print(resumo_uf)

################################################################################
#                    RESUMO POR TIPO DE ACIDENTE                               #
################################################################################

cat("\n=== RESUMO POR TIPO DE ACIDENTE ===\n")
resumo_tipo <- df_acidentes_prf %>%
  group_by(tipo_acidente) %>%
  summarise(
    Total_Acidentes = n(),
    Total_Obitos = sum(mortos, na.rm = TRUE),
    Total_Feridos = sum(feridos, na.rm = TRUE),
    Taxa_Mortalidade_pct = round(100 * sum(mortos, na.rm = TRUE) / n(), 2),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total_Acidentes))

print(resumo_tipo)

################################################################################
#                    RESUMO POR CLASSIFICAÇÃO DE ACIDENTE                      #
################################################################################

cat("\n=== RESUMO POR CLASSIFICAÇÃO ===\n")
resumo_classificacao <- df_acidentes_prf %>%
  group_by(classificacao_acidente) %>%
  summarise(
    Total_Acidentes = n(),
    Total_Obitos = sum(mortos, na.rm = TRUE),
    Total_Feridos = sum(feridos, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total_Acidentes))

print(resumo_classificacao)

################################################################################
#                    ANÁLISE TEMPORAL                                          #
################################################################################

cat("\n=== DISTRIBUIÇÃO POR DIA DA SEMANA ===\n")
resumo_dia_semana <- df_acidentes_prf %>%
  group_by(dia_semana) %>%
  summarise(
    Total_Acidentes = n(),
    Total_Obitos = sum(mortos, na.rm = TRUE),
    Total_Feridos = sum(feridos, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total_Acidentes))

print(resumo_dia_semana)

################################################################################
#                    VALIDAÇÃO DE INTEGRIDADE DOS DADOS                        #
################################################################################

cat("\n=== VERIFICAÇÃO DE INTEGRIDADE ===\n")

# Verificar se há IDs duplicados (devem ser únicos)
ids_duplicados <- df_acidentes_prf %>%
  group_by(id) %>%
  filter(n() > 1) %>%
  nrow()
cat("IDs duplicados:", ids_duplicados, "\n")

# Verificar consistência: total de feridos deve igualar (leves + graves)
inconsistencias_feridos <- df_acidentes_prf %>%
  filter(feridos != (feridos_leves + feridos_graves)) %>%
  nrow()
cat("Inconsistências de contagem de feridos:", inconsistencias_feridos, "\n")

# Verificar consistência: pessoas deve incluir (mortos + feridos + ilesos)
tolerancia <- 2  # Tolerância para pequenas diferenças
inconsistencias_pessoas <- df_acidentes_prf %>%
  mutate(total_calculado = mortos + feridos + ilesos + ignorados) %>%
  filter(abs(pessoas - total_calculado) > tolerancia) %>%
  nrow()
cat("Inconsistências de contagem de pessoas:", inconsistencias_pessoas, "\n")

# Verificar coordenadas geográficas válidas (Brasil)
coords_invalidas <- df_acidentes_prf %>%
  filter(latitude < -33 | latitude > 5 | longitude < -74 | longitude > -34) %>%
  nrow()
cat("Coordenadas geográficas inválidas:", coords_invalidas, "\n")

################################################################################
#                    CONFIRMAÇÃO DE CARGA FINALIZADA                           #
################################################################################

cat("\n✓ Dataset PRF carregado com sucesso!\n")
cat("   Total de acidentes:", nrow(df_acidentes_prf), "\n")
cat("   Total de óbitos:", sum(df_acidentes_prf$mortos, na.rm = TRUE), "\n")
cat("   Total de feridos:", sum(df_acidentes_prf$feridos, na.rm = TRUE), "\n")
cat("   Estados cobertos:", n_distinct(df_acidentes_prf$uf), "\n")
cat("   Rodovias federais:", n_distinct(df_acidentes_prf$br), "\n")
cat("   Período coberto:", min(df_acidentes_prf$data_inversa), "a", max(df_acidentes_prf$data_inversa), "\n\n")

# Salvar dados processados em RData para uso posterior
save(df_acidentes_prf, file = "Trabalho/bases-de-dados/acidentes/datatran2024.RData")
cat("✓ Dataset salvo em RData para carregamento futuro\n")
```
