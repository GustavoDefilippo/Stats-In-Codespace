O Polars é uma biblioteca moderna para análise de dados que surgiu em 2020, criada por Ritchie Vink como resposta às limitações de desempenho do Pandas, especialmente em grandes volumes de dados. O conceito central do Polars é oferecer uma engine de consulta analítica extremamente rápida, escrita em Rust e baseada no formato de memória columnar Apache Arrow, permitindo processamento paralelo e otimizações de memória.[1][2][3]

### Surgimento e Conceito

O Polars nasceu como um projeto pessoal para resolver gargalos de desempenho e uso de memória em bibliotecas tradicionais como o Pandas. Ele foi projetado para ser rápido, eficiente e escalável, especialmente para datasets grandes, utilizando técnicas de processamento paralelo e uma arquitetura close-to-the-metal, ou seja, próxima ao hardware. O Polars pode ser usado tanto de forma “eager” (execução imediata) quanto “lazy” (execução otimizada e diferida), permitindo que o usuário defina pipelines de transformação de dados que são otimizados antes da execução.[2][3][4][1]

### Vantagens do Polars

- **Desempenho**: O Polars é frequentemente 5 a 10 vezes mais rápido que o Pandas em operações comuns, podendo chegar a 30x em benchmarks específicos.[5][3][6]
- **Eficiência de memória**: Utiliza o formato columnar Apache Arrow, reduzindo o uso de memória e permitindo processar datasets maiores do que a RAM disponível, graças ao streaming e ao suporte a dados fora da memória.[3][2]
- **Paralelismo**: O Polars aproveita todos os núcleos do processador automaticamente, sem necessidade de configuração adicional.[2][3]
- **Sintaxe expressiva**: Oferece uma API intuitiva e encadeável, facilitando a leitura e manutenção do código.[7][4]
- **Integração com o ecossistema Python**: Permite fácil integração com outras bibliotecas populares, como Scikit-learn, Matplotlib e frameworks de machine learning.[7][3]
- **Open source e ativo**: O Polars é open source, possui uma comunidade crescente e está em constante evolução.[1][2]

### Exemplo concreto

Imagine que você está analisando dados de tráfego de uma cidade, com milhões de registros de veículos. Com o Pandas, operações como agrupamento, filtragem e agregação podem ser lentas e exigir muita memória. Já com o Polars, essas mesmas operações são executadas rapidamente, permitindo que você explore diferentes cenários e faça ajustes em tempo real, sem precisar esperar minutos ou horas para ver o resultado.[3][2]

### Conexão com o que já foi aprendido

```python
# Conversão de .tab para Polars (.parquet) e comparação de tempo e espaço com análise descritiva
# origem do banco de dados:https://dataverse.fgv.br/dataset.xhtml?persistentId=hdl:10438.3/FK2/S5ECCO
# Matrizes O-D detectadas a partir de duas chamadas de celular (RJ, Brasil) em 2014
# Instala Polars caso não esteja disponível
try:
    import polars as pl
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "polars>=1.0.0"])
    import polars as pl

import os
import time

# Caminhos de arquivo - tem que descomprimir o arquivo antes - gunzip "nome-arq"
tab_path = "/workspaces/Stats-In-Codespace/Aula1/julio/RJMA_ordinary_mobility_given_by_two_calls.tab"
parquet_path = os.path.splitext(tab_path)[0] + ".parquet"

def human_bytes(n: int) -> str:
    for unit in ["B","KB","MB","GB","TB"]:
        if n < 1024:
            return f"{n:.2f} {unit}"
        n /= 1024
    return f"{n:.2f} PB"

# 1) Ler .tab com Polars e análise descritiva (medir tempo e memória)
t0 = time.perf_counter()
df_tab = pl.read_csv(
    tab_path,
    separator="\t",
    infer_schema_length=10000,
    low_memory=True,
    null_values=["", "NA", "NaN", "null", "NULL"]
)
t_read_tab = time.perf_counter() - t0

mem_tab = df_tab.estimated_size()  # bytes

t0 = time.perf_counter()
desc_tab = df_tab.describe()
t_desc_tab = time.perf_counter() - t0

# 2) Converter para Parquet (formato coluna eficiente) e medir espaço/tempo
t0 = time.perf_counter()
df_tab.write_parquet(parquet_path, compression="zstd", statistics=True)
t_write_parquet = time.perf_counter() - t0

size_tab = os.path.getsize(tab_path)
size_parquet = os.path.getsize(parquet_path)

# 3) Ler Parquet e fazer a mesma análise (medir tempo e memória)
t0 = time.perf_counter()
df_parquet = pl.read_parquet(parquet_path)
t_read_parquet = time.perf_counter() - t0

mem_parquet = df_parquet.estimated_size()

t0 = time.perf_counter()
desc_parquet = df_parquet.describe()
t_desc_parquet = time.perf_counter() - t0

# 4) Resultados
print("Resumo de Tamanhos em Disco:")
print(f"- .tab:      {human_bytes(size_tab)}")
print(f"- .parquet:  {human_bytes(size_parquet)}")
if size_tab > 0:
    print(f"- Fator de redução: {size_tab/size_parquet:.2f}x (tab/parquet)")

print("\nTempos (segundos):")
print(f"- Leitura .tab:          {t_read_tab:.4f}s")
print(f"- Escrita .parquet:      {t_write_parquet:.4f}s")
print(f"- Leitura .parquet:      {t_read_parquet:.4f}s")
print(f"- Describe .tab:         {t_desc_tab:.4f}s")
print(f"- Describe .parquet:     {t_desc_parquet:.4f}s")

print("\nMemória estimada dos DataFrames:")
print(f"- DF (.tab):      {human_bytes(mem_tab)}")
print(f"- DF (.parquet):  {human_bytes(mem_parquet)}")

print("\nEsquema (dtypes) detectado:")
print(df_tab.schema)
```

```python
import polars as pl
import matplotlib.pyplot as plt

# Carregar os dados (assumindo que o caminho está correto)
od2calls = "/workspaces/Stats-In-Codespace/Aula1/julio/RJMA_ordinary_mobility_given_by_two_calls.parquet"
df_od2calls = pl.read_parquet(od2calls)

# Calcular o total geral de viagens para o Pareto
total_all = df_od2calls['travel_count_fix_factor'].sum()

# Agrupar por 'destination' e somar 'travel_count_fix_factor'
freq_viagens = df_od2calls.group_by('destination').agg(
    pl.col('travel_count_fix_factor').sum().alias('total_travel_count')
).sort('total_travel_count', descending=True).head(10)

# Calcular a soma cumulativa e a porcentagem cumulativa
cumulative = freq_viagens.select([
    pl.col('destination'),
    pl.col('total_travel_count'),
    pl.col('total_travel_count').cum_sum().alias('cumulative_sum'),
    (pl.col('total_travel_count').cum_sum() / total_all * 100).alias('cumulative_pct')
])

# Extrair listas para plotagem
destinos = cumulative['destination'].to_list()
total_travel = cumulative['total_travel_count'].to_list()
cum_pct = cumulative['cumulative_pct'].to_list()

# Criar o gráfico de Pareto
fig, ax1 = plt.subplots(figsize=(14, 7))  # Aumentei o tamanho da figura para mais espaço

# Eixo primário: barras para total de viagens
bars = ax1.bar(destinos, total_travel, color='cornflowerblue', alpha=0.7, edgecolor='navy')
ax1.set_xlabel('Destino', fontsize=14)
ax1.set_ylabel('Total de Viagens', fontsize=14, color='cornflowerblue')
ax1.tick_params(axis='y', labelcolor='cornflowerblue')
ax1.set_title('Gráfico de Pareto: Top 10 Destinos por Número Total de Viagens', fontsize=16)

# Eixo secundário: linha para porcentagem cumulativa
ax2 = ax1.twinx()
line = ax2.plot(destinos, cum_pct, color='red', marker='o', linewidth=2, markersize=6)
ax2.set_ylabel('Porcentagem Cumulativa (%)', fontsize=14, color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, max(cum_pct) * 1.1)  # Ajustar limite superior

# Ajustes para o eixo X: rotação de 45 graus (já tinha), alinhamento à direita, e fonte menor
plt.xticks(rotation=45, ha='right', fontsize=10)  # Diminuí a fonte para 10

# Adicionar linha de referência em 80% (regra de Pareto clássica)
ax2.axhline(y=80, color='orange', linestyle='--', alpha=0.7, label='Linha de 80%')

# Legenda
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2 + line, labels1 + labels2 + ['Pareto (80%)'], loc='upper left')

plt.tight_layout()
plt.show()

# Opcional: Imprimir os dados para verificação
print(cumulative)
```

### O que significa visualizar um gráfico em escala log10?

Quando utilizamos uma escala logarítmica (log10) em um gráfico, estamos transformando os valores dos eixos para que sejam representados como potências de 10. Isso significa que:

- Cada unidade no eixo representa um aumento exponencial (multiplicação por 10).
- Por exemplo, os valores 1, 10, 100 e 1000 seriam igualmente espaçados no gráfico.

#### Por que usar escala logarítmica?
1. **Ampliar pequenos valores e comprimir grandes valores**: Isso é útil quando os dados possuem uma grande variação (valores muito pequenos e muito grandes).
2. **Identificar padrões em dados exponenciais**: Dados que crescem ou decrescem exponencialmente aparecem como linhas retas em uma escala logarítmica.
3. **Melhorar a visualização**: Facilita a análise de dados distribuídos de forma desigual, destacando tendências em diferentes ordens de magnitude.

#### Exemplo prático:
Se os dados representam populações de cidades e algumas têm centenas de habitantes enquanto outras têm milhões, uma escala log10 permite visualizar todas as cidades no mesmo gráfico sem que as menores sejam "esmagadas" pelos valores maiores.

```python
# Gerar stem-and-leaf plot em modo texto
data = df_od1pop['census_population'].to_numpy()

# Ordenar os dados
data_sorted = np.sort(data)

# Gerar os "stems" e "leaves"
stems = np.floor(data_sorted / 10).astype(int)
leaves = (data_sorted % 10).astype(int)

# Agrupar os valores por "stem"
stem_leaf_dict = {}
for stem, leaf in zip(stems, leaves):
    if stem not in stem_leaf_dict:
        stem_leaf_dict[stem] = []
    stem_leaf_dict[stem].append(leaf)

# Exibir o stem-and-leaf plot
print("Stem-and-Leaf Plot: Census Population")
for stem, leaf_list in stem_leaf_dict.items():
    leaf_str = " ".join(map(str, leaf_list))
    print(f"{stem} | {leaf_str}")
```

```python
#Mostrar o formato da distribuição da variável 'census_population' do DataFrame df_od2calls
# usar 1o coeficiente de pearson para mostrar assimetria
import polars as pl
od1call = "/workspaces/Stats-In-Codespace/Aula1/julio/RJMA_ordinary_visiting_given_by_one_call_over_shifts.parquet"
df_od1call = pl.read_parquet(od1call)
# Calcular estatísticas descritivas
mean = df_od1call.select(pl.col('census_population').mean()).item()
median = df_od1call.select(pl.col('census_population').median()).item()
std_dev = df_od1call.select(pl.col('census_population').std()).item()

# Verificar se os valores não são None antes de calcular
if mean is not None and median is not None and std_dev is not None and std_dev != 0:
    pearson_skewness = 3 * (mean - median) / std_dev
else:
    pearson_skewness = None
    print("Não foi possível calcular o coeficiente de assimetria (valores None ou desvio padrão zero)")
print(f"Coeficiente de Assimetria de Pearson para 'travel_count_fix_factor': {pearson_skewness:.4f}")
# Interpretar o resultado
if pearson_skewness > 0:
    print("A distribuição é assimétrica à direita (cauda longa à direita).")
elif pearson_skewness < 0:
    print("A distribuição é assimétrica à esquerda (cauda longa à esquerda).")
else:
    print("A distribuição é simétrica.")
```

```python
#curtose: usar a formula de fischer para ver a curtose da variável 'census_population' do DataFrame df_od1call
import polars as pl
od1call = "/workspaces/Stats-In-Codespace/Aula1/julio/RJMA_ordinary_visiting_given_by_one_call_over_shifts.parquet"
df_od1call = pl.read_parquet(od1call)
# Calcular estatísticas descritivas
kurtosis = df_od1call.select(pl.col('census_population').kurtosis()).item()
print(f"Curtose de Fisher para 'census_population': {kurtosis:.4f}")
# Interpretar o resultado
if kurtosis > 0:
    print("A distribuição é leptocúrtica (Distribuição alongada, com pico alto e caudas pesadas (maior concentração na média, mais outliers), curtose positiva.).")
elif kurtosis < 0:
    print("A distribuição é platicúrtica (Distribuição achatada, com pico baixo e caudas leves, curtose negativa.).")
else:
    print("A distribuição é mesocúrtica (similar à normal).")
```

```python
# Tabela de contingência entre as variáveis 'presumed_domicile' e 'destination' do data frame df_expanded
# Usar heatmap para visualizar a tabela de contingência
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt

# Cria a tabela de contingência usando Polars
contingency_table = df_expanded.group_by(["presumed_domicile", "destination"]).agg(
    pl.count().alias("count")
).pivot(index="presumed_domicile", columns="destination", values="count").fill_null(0)

# Converte para pandas e define 'presumed_domicile' como índice para o heatmap
contingency_pd = contingency_table.to_pandas()
contingency_pd = contingency_pd.set_index('presumed_domicile')

# Plota a tabela de contingência SEM anotações (apenas cores)
plt.figure(figsize=(14, 10))
plt.title("Tabela de Contingência entre 'presumed_domicile' e 'destination'")
sns.heatmap(contingency_pd, annot=False, cmap="YlGnBu", cbar_kws={'label': 'Contagem'})
plt.xlabel("Destination")
plt.ylabel("Presumed Domicile")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# teste do qui-quadrado de independência entre as variáveis 'presumed_domicile' e 'destination' do data frame df_expanded
# usando scipy.stats.chi2_contingency
import polars as pl
from scipy.stats import chi2_contingency

# Cria a tabela de contingência entre 'presumed_domicile' e 'destination'
contingency_table = df_expanded.group_by(["presumed_domicile", "destination"]).agg(
    pl.count().alias("count")
).pivot(index="presumed_domicile", columns="destination", values="count").fill_null(0)

# Converte a tabela de contingência para um array NumPy
contingency_array = contingency_table.select(pl.all().exclude("presumed_domicile")).to_numpy()

# Realiza o teste do qui-quadrado
chi2, p, dof, expected = chi2_contingency(contingency_array)

# Exibe os resultados
print("=== Teste Qui-Quadrado de Independência ===")
print(f"Estatística Qui-Quadrado: {chi2:.4f}")
print(f"Valor-p: {p:.10f}")
print(f"Graus de Liberdade: {dof}")
print("\nFrequências Esperadas (primeiras 5x5 para visualização, se a tabela for grande):")
print(expected[:5, :5] if expected.shape[0] > 5 and expected.shape[1] > 5 else expected)

# Interpretação básica
alpha = 0.05
if p < alpha:
    print("\nConclusão: Há evidência significativa de dependência entre 'presumed_domicile' e 'destination' (p < 0.05).")
else:
    print("\nConclusão: Não há evidência significativa de dependência entre 'presumed_domicile' e 'destination' (p >= 0.05).")
```
